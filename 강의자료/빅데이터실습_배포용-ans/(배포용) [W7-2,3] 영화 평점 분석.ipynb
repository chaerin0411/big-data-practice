```python
import pandas as pd
from pandas import Series, DataFrame
import numpy as np
```

## 1. 영화 평점 데이터 적재 및 전처리

```python
# 사용자 데이터 읽어오기
users = pd.read_csv('data/movielens/users.dat', sep = '::', engine = 'python',
                   names = ['사용자아이디', '성별','연령','직업','지역'])
users.head()
```

```python
# 평점 데이터 읽어오기
ratings = pd.read_csv('data/movielens/ratings.dat', sep = '::', engine = 'python',
                   names = ['사용자아이디', '영화아이디','평점','타임스탬프'])
ratings.head()
```

```python
# 영화데이터 읽어오기
movies = pd.read_csv('data/movielens/movies.dat', sep = '::', engine = 'python',
                   names = ['영화아이디','영화제목','장르'], encoding = 'latin-1')
movies.head()
```

```python
# 3개의 데이터프레임을 하나로 합치기
# ratings 데이터를 기반으로 영화아이디, 영화제목을 같이 봐야 할 경우
# 사용자아이디를 users의 성별, 나이 정보를 같이 평점을 분석할 경우
# merge : 사용자아이디를 기준으로 합침
# data에 users, ratings 2개의 dataframe을 합침
# on 인자를 사용하지 않아도 사용자아이디 공통 columns이 있으므로 생략 가능
data = pd.merge(users, ratings)
data.head()
```

```python
# data에 추가로 movies의 data를 합침
# on 인자를 사용하지 않아도 영화아이디 공통 columns이 있으므로 생략 가능
data = pd.merge(data, movies)
data.head()
```

```python
# 영화들의 평점 평균을 구하여, 평점이 높은 영화 찾기 
# data.평점.mean() : 전체 영화, 전체 사용자 평점
# 영화별 평점은 그룹 집계(pivot_table) 사용
# 동일한 영화제목끼리 평점의 평균(mean)을 구함
영화평점 = data.pivot_table(index = '영화제목', aggfunc = 'mean', values = '평점')
```

```python
영화평점.sort_values(by = '평점', ascending = False)
```

## [실습 #1] 여자들이 좋아하는 영화 찾기 
### - 여성 평점이 4.0 이상이고 여성 평점의 개수가 500개 이상인 영화

```python
# 여자들이 좋아한는 영화를 찾는 법
# 01. 모든 영화 평점 데이터에서 여성 평점 데이터만 선택하여 영화별로 평점의 평균과 개수를 구함
# 02. 성별로 평점을 각각 구해서 여성 평점만 취함
```

```python
# 영화별 성별 평점 구하기
# pivot_table에 index='영화제목', columns='성별', 평점에 대한 평균과 개수 구함
# 평균의 평점과 개수가 각각 female과 male로 남성과 여성으로 각각 나뉘어서 데이터가 저장
ex1 = data.pivot_table(index = '영화제목', columns = '성별', aggfunc = ['mean', 'count'], values = '평점')
ex1
```

```python
# 조건 색인
# 튜플의 형태로, (mean, '여성'), 즉 여성평점의 평균을 선택
# 4.0 이상인 경우만 True가 나옴
# 튜플의 형태로, (count, '여성'), 즉 여성평점의 개수를 선택
# 500개 이상인 경우만 True가 나옴
# 2개를 모두 만족하는 경우만 True가 나옴
# 여성인기영화로 저장
여성인기영화 = ex1[(ex1[('mean', 'F')] >= 4.0) & (ex1[('count', 'F')] >= 500)]
```

```python
여성인기영화
```

## [실습 #2] 실습 #1에서 구한 영화(여성인기영화)의 장르를 분석해 보자.
여성인기영화의 장르 통계 구하기

예를 들어, 여성인기영화 중 Drama 장르의 영화는 10개, Action 영화는 3개, ...

```python
# 장르 정보는 movies에 존재
# 전체 영화에 대한 장르가 아닌 여성인기영화에 해당하는 장르만 필요
# isin()을 사용하여 movies에서 여성인기영화의 장르만 선택
# data.isin() : data의 값이 인자로 받은 리스트에 포함되어 있는 경우에만 True
# 영화아이디 1,3,5,9만 조건색인으로 넣어줌
movies[movies.영화아이디.isin([1,3,5,9])]
```

```python
# 영화아이디를 영화제목으로 변경
# 1,3,5,9 대신 여성인기영화.index로 바꿈
# 장르 정보만 선택
# 여성인기영화장르로 저장
# 하나의 영화가 여러 장르에 속해 있는 경우, 세로줄로 구분되어 있음
여성인기영화장르 = movies[movies.영화제목.isin(여성인기영화.index)].장르
```

```python
# str.split() 함수를 활용하여 장르 정보를 구분자로 구분
# 여성인기영화장르로 저장
여성인기영화장르 = 여성인기영화장르.str.split('|', expand = True)
```

```python
여성인기영화장르
```

```python
# 영화장르 카운트하기
# 컬럼별로 value_count()를 수행한 결과값들을 모두 더함
# 동일한 인덱스가 두 군데 모두 있는 경우에만 정상적으로 산술연산이 수행되고, 둘 중 하나라도 인덱스가 없다면 NaN값으로 처리됨
여성인기영화장르[0].value_counts() + 여성인기영화장르[1].value_counts()
```

```python
# +를 add로 변경
# fill_value = 0
여성인기영화장르[0].value_counts().add(여성인기영화장르[1].value_counts(), fill_value = 0)
```

```python
# for문 활용
# 여성인기장르를 Series로 생성
# 5개의 columns 순회
# '여성인기장르'에 5번을 더함
# 더할 때 index로 col을 넣어줌
# DeprecationWarning : series 생성 시 object로 기본 데이터 타입이 설정되므로 float하기 위해서는 float로 변경하라는 뜻
# DeprecationWarning을 없애기 위해서는 dtype을 float64로 지정
# 여성인기장르를 더한 값을 여성인기장르에 할당하여 누적
여성인기장르 = Series(dtype = 'float64')
for col in 여성인기영화장르.columns:
    여성인기장르 = 여성인기장르.add(여성인기영화장르[col].value_counts(), fill_value = 0)
```

```python
여성인기장르.sort_values(ascending = False)
```

## [실습 #3] 남자와 여자의 호불호가 크게 갈리는 영화 10개 찾기
전체 평점의 개수가 500개 이상인 영화만 대상으로 함.

```python
# 영화별 성별 평점 정보를 ex3에 저장
ex3 = data.pivot_table(index = '영화제목', columns = '성별', aggfunc = ['mean', 'count'], values = '평점')
ex3
```

```python
# 전체 평점 개수가 500개 이상인 영화 선택
# 여성 평점 개수와 남성 평점 개수의 합이 500이상인 영화 선택
# 조건 색인 수행
# ex3_500에 저장
ex3_500 = ex3[ex3[('count', 'F')] + ex3[('count', 'M')] >= 500]
ex3_500
```

```python
# ex3_500을 대상으로 남자와 여자의 호불호가 크게 갈리는 영화 찾기
# 남자평점과 여자평점의 차이를 계산하여 차이가 많이 나는 것을 선택
# abs() 함수를 통해 차이를 절댓값으로 표현
# 평점 차이의 절댓값을 평점차이 컬럼으로 추가
# SettingWithCopyWarning : loc를 사용하여 접근하라는 의미
# 3군데 모두.loc를 넣고 :,를 넣어줌
ex3_500.loc[:, '평점차이'] = abs(ex3_500.loc[:, ('mean', 'F')] - ex3_500.loc[:, ('mean', 'M')])
```

```python
ex3_500
```

```python
# 호불호가 크게 갈리는 영화 10개 찾기
# 평점의 차이가 큰 10개를 찾기 위해 sort_values()로 정렬
ex3_500.sort_values(by = '평점차이', ascending = False).head(10)
```

```python
# nlargest() 사용 가능
ex3_500.nlargest(10, '평점차이')
```

## [실습 #4] 연령대 별로 영화 평점 분석하기
연령대(10대 미만, 10대, 20대, ...50대) 컬럼을 추가한 후, 영화별 연령대별 영화평점 구하기

```python
# data에는 연령 정보만 있고, 연령대 정보는 없음
# 연령 정보에서 10살 단위로 구분하여 연령대라는 새로운 columns을 만듬
# 연령 columns의 값 분포 확인
# describe() 함수 활용
# 10대 미만, 10대, 20대, 30대, 40대, 50대 이상
data.describe()
```

```python
# 연령대 columns 추가하는 방법
# 01. 함수 사용
# generate_ages(x): 연령값을 인자 x로 받아 연령대를 리턴하는 함수('10대 미만' ~ '50대 이상')
# x가 10보다 작으면 10대 미만
# x가 10보다 크고 20보다 작으면 10대
# x가 20보다 크고 30보다 작으면 20대
# x가 30보다 크고 40보다 작으면 30대
# x가 40보다 크고 50보다 작으면 40대
# x가 50보다 크면 50대 이상
def generate_ages(x):
    if x < 10:
        return '10대 미만'
    elif x < 20:
        return '10대'
    elif x < 30:
        return '20대'
    elif x < 40:
        return '30대'
    elif x < 50:
        return '40대'
    else:
        return '50대 이상'
```

```python
# generate_ages(35)를 실행하면 30대 출력
generate_ages(35)
```

```python
# generate_ages()를 연령 컬럼에 적용
# generate_ages()에 data.연령을 전달하는 경우 에러 발생
# generate_ages()는 인자가 숫자값이어야 하지만, Data.연령은 Series이기 때문에 데이터 타입이 맞지 않음
# data.연령의 각 값들에 함수를 적용하기 위해서 apply() 함수를 사용
# 함수의 결과를 연령대 컬럼으로 추가
data['연령대'] = data.연령.apply(generate_ages)
```

```python
# Series 생성(sr)
# np.digitize 실행
# np.digitize()
# Series에 해당하는 값들을 두 번째 인자값으로 구분하여 특정 숫자값으로 변환함
# 구분하는 기준은 두 번째 인자값
sr = Series([13, 15, 45, 53, 1, 63])
np.digitize(sr, [10, 20, 30])
```

```python
# map()
# 0,1,2,3,4,5을 10대미만, 10대, 20대, 30대, 40대, 50대 이상으로 맵핑하여 변환
# 1:1로 값을 치환할 때 사용
# 0~5 값이 원하는 문자열로 변환됨
# data['연령대2']로 다시 할당
data['연령대2'] = data.연령대2.map({
    0: '10대 미만',
    1: '10대',
    2: '20대',
    3: '30대',
    4: '40대',
    5: '50대 이상'
})
```
