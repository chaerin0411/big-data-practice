1.1.4 두 개 그래프를 서로 다른 y축 적용하기

```python
sr2 = Series([500,300,250,180,700], index = ['A','B','C','D','E'])
sr2
```

```python
# plt.subplots으로 새로운 figure와 새로운 axes 생성
# sr : line graph를 빨간색으로 그림
# sr2 : bar graph 그림
fig, axes = plt.subplots()
axes.plot(sr, color = 'r')
axes.bar(sr2.index, sr2.values)
```

```python
# sr과 sr2의 y값의 범위가 차이가 많이 나서 재대로 보이지 않음
# 시리즈 값이 상대적으로 작아 값의 변화를 이해하기 어려움
# 2개의 y축을 각각 적용
# 왼쪽 : bar graph, 오른쪽 line graph
# y축이 다르기 때문에 서로 다른 axes에 각각 그려주어야 함
# axes2 = axes.twinx() : x축은 공유
# 첫 번째 axes와 두 번째 axes는 x축을 공유해서 중첩하겠다는 의미
fig, axes = plt.subplots()
axes.plot(sr, color = 'r')
axes2 = axes.twinx()
axes2.bar(sr2.index, sr2.values)
```

```python
# tick_params를 사용하여 line graph y축을 라벨컬러를 레드로 설정
# alpha값으로 투명도 조절(0~1 사이의 값)
fig, axes = plt.subplots()
axes.plot(sr, color = 'r')
axes.tick_params(axis = 'y', labelcolor = 'r')
axes2 = axes.twinx()
axes2.bar(sr2.index, sr2.values, alpha = 0.3)
```

### 1.2. 여러 개 그래프 그리기 
하나의 figure 안에 여러 개의 axes를 그리기

```python
# plt.subplots(2,2)로 총 4개 생성
# axes_list에는 4개의 각각의 axes가 2X2 배열 형태로 저장되어 있음
fig, axes_list = plt.subplots(2,2)
```

```python
import random
```

```python
# plt.subplots(2,2)로 총 4개 생성
# axes_list에는 4개의 각각의 axes가 2X2 배열 형태로 저장되어 있음
# axes_list[0,0]에는 sr를 plot으로 그림
# axes_list[0,1]에는 히스토그램(histogram) 그림
# 랜덤하게 생성할 값의 범위 : 1~100 사이 값으로 50개 생성
# bins : 5개의 그룹으로 나눔
# 5개의 버켓으로 데이터의 분포도를 확인 가능
# axes_list[1,0]에는 Bar graph 그림
# axes_list[1,1]에는 scatter graph 그림
# scatter graph는 2차원 데이터의 분포로 2개의 값이 필요함
# random.sample을 x값과 y값에 적용
# 알파값을 적용하여 데이터의 밀집도를 표시함
fig, axes_list = plt.subplots(2,2)
axes_list[0,0].plot(sr)
axes_list[0,1].hist(random.sample(range(1,100), 50), bins = 5, color = 'r', alpha = 0.3)
axes_list[1,0].bar(sr2.index, sr2.values)
axes_list[1,1].scatter(random.sample(range(1,100), 90), random.sample(range(1,100), 90), alpha = 0.3)
```

```python
# axes_list[0,0]에 set_title로 axes의 이름 부여
# axes_list[0,0]에 x, y label 붙이기
# x축 제목은 아래쪽 그래프(axes에 의해 가려져서 안보이는 것임)
# 각 axes들의 여백 조정 plt.subplots_adjust(hspace = 0.5, wspace = 0.3)
fig, axes_list = plt.subplots(2,2)
axes_list[0,0].plot(sr)
axes_list[0,0].set_title('Axes #1')
axes_list[0,0].set_xlabel('x1')
axes_list[0,0].set_ylabel('y1')
axes_list[0,1].hist(random.sample(range(1,100), 50), bins = 5, color = 'r', alpha = 0.3)
axes_list[1,0].bar(sr2.index, sr2.values)
axes_list[1,1].scatter(random.sample(range(1,100), 90), random.sample(range(1,100), 90), alpha = 0.3)
plt.subplots_adjust(hspace = 0.5, wspace = 0.3)
```

### 1.3. <a href = https://matplotlib.org/3.1.1/tutorials/introductory/usage.html#what-is-a-backend> backend </a> 개념
Matplotlib은 매우 다양한 usecase에 사용되며, 이를 모두 충족시키기 위하여 다양한 backend를 지원함.  
가장 많이 활용하는 옵션은 inline으로, 그래프를 output창에 바로 표현되도록 하는 옵션이며, <strong>%maptlotlib inline</strong>를 실행하여 설정한다. 


%로 시작하는 명령어는 Cell Magic 명령어로써, 파이썬 코드가 아니라 Jupyter에게 특정 기능을 수행하도록 하는 명령이다. 즉, %matplotlib inline 명령은 Jupyter에게 matplotlib 그래프를 출력 영역에 표시할 것을 지시하는 명령이다.

```python
%matplotlib inline
```

#### 4) (실습 #1) 가문 별로 전쟁을 참여한 횟수 시각화

```python
# battles : 전쟁에 대한 정보
# 가문 정보는 총 8개 컬럼으로 나뉘어 제공
# 필요한 컬럼만 선택 : attacker_1 ~ defender_4(총 8개 컬럼) 선택
# 01. attacker_1 ~ defender_4까지 모두 나열
# 가장 기본적인 방식으로 8개 컬럼을 모두 나열
# 로우 인덱스 0번 Lannister 가문이 Tully 가문 공격
# 로우 인덱스 14번 Stark와 Tully 가문이 합동으로 Lannister 가문 공격
# 로우 인덱스 27번 Free folk와 Thenns, Giants 합동 공격 / Night's Watch와 Baratheon 합동 수비
battles[['attacker_1', 'attacker_2', 'attacker_3', 'attacker_4', 'defender_1', 'defender_2', 'defender_3', 'defender_4', ]]
```

```python
# 02. loc를 활용한 컬럼 슬라이싱
# 8개 컬럼이 연속되어 있으므로, 컬럼명을 슬라이싱하여 얻을 수 있음
battles.loc[:, 'attacker_1':'defender_4']
```

```python
# 03. columns를 활용한 컬럼명 슬라이싱
# battles.columns : 컬럼 목록 구한 후 색인
# Columns를 슬라이싱하여 얻은 컬럼 목록을 색인기호에 넣어줌
battles[battles.columns[5:13]]
```

```python
# 04. 정규 표현식 활용
# 패턴을 가지고 있는 경우에는 정규 표현식을 사용할 수 있음
# 정규 표현식은 컬럼의 연속 유무와 상관없이 활용 가능
# battles.filter() : regex 인자값 설정 -> 특정 패턴 가진 컬럼명만 선택 가능
battles_house = battles.filter(regex = 'attacker_[1-4]|defender_[1-4]')
```

```python
# 각 가문들이 출현한 횟수를 집계
# value_counts() : 각 가문의 출연 빈도수 세기
battles_house.attacker_1.value_counts()
```

```python
battles_house.attacker_2.value_counts()
```

```python
battles_house.attacker_4.value_counts()
```

```python
# 같은 인덱스 라벨끼리 더해지면서 둘 중 하나라도 없으면 NaN값으로 표시됨
battles_house.attacker_1.value_counts() + battles_house.attacker_2.value_counts()
```

```python
# fill_value = 0
# 값이 없는 경우 0으로 처리
battles_house.attacker_1.value_counts().add(battles_house.attacker_2.value_counts(), fill_value = 0)
```

```python
# house_to_battles_count 변수 : 누적된 값 저장
# DeprecatedWarning : Default 값이 바뀌었음을 의미
# Default가 'object'에서 'float64'로 바뀌니 dtype을 명시하라는 의미
house_to_battles_count = Series(dtype = 'float64')
# 컬럼들을 각각 순회
for col in battles_house.columns:
     house_to_battles_count = house_to_battles_count.add(battles_house[col].value_counts(), fill_value = 0)
```

```python
# house_to_battles_count를 그래프로 시각화
# x축이 범주형 데이터인 경우에는 막대 그래프가 더 적합
# 집계를 정렬해서 보고 싶은 경우, plot 함수 수행 전에 정렬
# sort_values() : 오름차순 정렬
house_to_battles_count.sort_values().plot(kind = 'barh')
```

#### 5) (실습 #2) 년도별로 사망자의 숫자와 전쟁이 벌어진 횟수 시각화

```python
# 01. 연도별 사망자 수 구하기
# deaths['Death Year'].value_counts() : 연도별 사망자 수
deaths_for_year = deaths['Death Year'].value_counts().sort_index()
```

```python
# 02. 연도별 전쟁 횟수 구하기
# battles.year.value_counts() : 연도별 전쟁 횟수
# pivot_table 사용
# aggfunc을 적용할 values는 null값이 없는 컬럼으로 선택
# 자동 정렬되어 출력
# values의 null 값이 없는 컬럼을 찾아야 함
battles_for_year = battles.pivot_table(index = 'year', aggfunc = 'count', values='name')
```

```python
battles.info()
```

```python
# 03. 두 분석 결과를 하나의 데이터프레임으로 합치기
deaths_for_year
```

```python
battles_for_year
```

```python
# pd.concat : 데이터프레임 합치기
# 컬럼 방향으로 합칠 것이기 때문에 axis=1로 지정
data = pd.concat([deaths_for_year, battles_for_year], axis=1)
data
```

```python
# 컬럼명 변경
data.columns = ['# of deaths', '# of battles']
```

```python
# fillna() : 결측값 채움
data.fillna(0, inplace = True)
```

```python
# 연도별 전쟁 횟수와 사망자 수 시각화
data.plot(figsize = (5,3),
         title = '연도별 사망자와 전쟁횟수 추이',
         xticks = data.index,
         xlabel = '연도',
         ylabel = '횟수')
```

```python
# 연도별 전쟁 횟수와 사망자 수의 상광 관계 분석
# corr() 상관 관계 분석
# 상관값이 1에 가까울수록 상관 관계가 높다는 의미
data.corr()
```
