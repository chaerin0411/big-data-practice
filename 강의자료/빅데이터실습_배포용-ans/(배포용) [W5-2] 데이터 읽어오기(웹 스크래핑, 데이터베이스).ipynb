```python
# read_html() 응용하기
# 01.1~50페이지 중 10개 페이지만 데이터 읽어오기
# 02.10개 페이지를 하나의 데이터로 합치기
# 03.URL의 페이지 번호를 실행하기
# https://finance.naver.com/sise/sise_market_sum.naver?&page=10
```

```python
# 시가총액이 높은 상위 500개 기업 데이터 읽어오기
# 01.페이지 번호를 제외한 공통 부분을 url에 저장
url = 'https://finance.naver.com/sise/sise_market_sum.naver?&page='
```

```python
# 02.페이지 번호를 1~10까지 변경하면서 10번 반복(Loop)하여 데이터 가져오기
# num에 1~10까지 할당이 되면서 10번을 순회하게 됨
# 03. 코스피500이라는 변수에 read.html()의 결과를 저장
코스피500 = []
for num in range(1,11):
    full_url = url + str(num)
    data = pd.read_html(full_url, encoding = 'cp949')
    df = data[1]
    df = df[df.종목명.notnull()]
    코스피500.append(df)
```

```python
# 항목당 50개의 데이터씩 총 500개 기업에 해당하는 데이터를 가져온 것을 확인할 수 있음
코스피500[0]
```

```python
코스피500[1]
```

```python
코스피500[9]
```

```python
data = pd.concat(코스피500)
```

```python
# HTML 문서로도 변환하기
# 코스피 시가총액 상위 500개 기업 데이터가 저장된 HTML 문서가 생성
data.to_html('data/코스피500데이터.html', encoding = 'cp949')
```

### 3) Selenium 라이브러리 활용  

* selenium 설치: conda install selenium

```python
# selenium을 활용한 동적 웹페이지 스크래핑
from selenium import webdriver
```

```python
# webdriver를 사용해 크롬 브라우저를 활용하기
# Chrome() 함수의 인자에 chromedriver 파일의 경로를 지정
driver = webdriver.Chrome('chromedriver')
```

```python
# 데이터 로딩할 시간을 줌
driver.implicitly_wait(3)
```

```python
# Selenium 라이브러리가 이 URL을 실행하면 자동으로 이 페이지를 띄워서 접근하는 것을 확인해 볼 수 있음
driver.get(삼성전자url)
```

```python
# selenium에 의해서 수행된 웹 사이트의 html 문서가 저장됨
html = driver.page_source
```

```python
# 저장된 html 문서를 read_html() 문서로 읽어 봄
data = pd.read_html(html, encoding = 'cp949')
```

```python
data[0]
```

```python
삼성_일별 = data[0][data[0].날짜.notnull()]
삼성_일별
```

```python
# 실습. 모든 기간의 삼성전자 일별 데이터를 스크래핑 해 하나의 데이터프레임으로 저장하기
일별주가 = []
삼성전자url = 'https://finance.naver.com/item/sise_day.naver?code=005930&page='

for num in range(1, 200): # for num in range(1, 619)
    full_url = 삼성전자url + str(num)
    driver.get(full_url)
    html = driver.page_source
    data = pd.read_html(html, encoding = 'cp949')
    삼성_일별 = data[0][data[0].날짜.notnull()]
    일별주가.append(삼성_일별)
```

```python
삼성전자_일별_최종데이터 = pd.concat(일별주가)
```

## 4. 데이터베이스에서 읽기
### 파이썬에 내장(embedded)된 sqlite 사용

```python
# 데이터베이스
# 데이터를 관리하기 위해서 사용하는 프로그램으로써, 데이터베이스관리시스템(DBMS)이라고도 함
# sqlite를 활용한 데이터 읽고 쓰기
# sqlite 데이터베이스는 anaconda를 설치할 때 함께 설치됨
```

```python
# 01.sqlite3 라이브러리 import 하기
import sqlite3
```

```python
# 02.SQL 문법에 맞춰 테이블을 생성하는 구문 만들기
#테이블 생성 SQL 정의
# a,b,c 세 개의 값을 가진 locations 테이블을 생성하는 구문
# 칼럼들의 데이터 타입
테이블생성 = 'CREATE TABLE locations \
                (a varchar[20], b varchar[20], c integer)'
```

```python
# 03.sqlite에 접속해 구문 실행하기
# sqlite 접속
# sqlite3.connect한 후 데이터베이스 명을 주어 conn 변수에 저장함
# sqlite에 접속하기 위한 커넥션 객체를 conn에 저장
conn = sqlite3.connect('temp.db')
```

```python
#테이블 생성 SQL 수행 
conn.execute(테이블생성)
```

```python
# commit() 수행. - SQL 수행 후, commit을 해야만, 데이터베이스에 SQL 구문이 적용됨 
conn.commit()
```

```python
# 테이블을 생성하는 행위는 엑셀로 비유하면, 파일(혹은 sheet)을 새로 만들고 이름을 locations라고 하고, 
# a,b,c 컬럼을 저장하는 틀을만든 것과 같다.
```

```python
#생성된 테이블에 저장할 데이터 정의
data = [['서울','강남', 15]
       ,['서울','강동', 10]
       ,['서울','동대문', 7]
       ,['서울','양천', 12]
       ,['성남','분당', 12]
       ,['수원','영통', 15]
       ,['성남','중원', 11]
       ,['수원','팔달', 14]
       ]
```

```python
# 데이터 적재 SQL 정의 
# 생성된 테이블에 값 삽입하기
# 데이터 하나를 삽입하는 SQL 구문 : insert into
삽입 = 'insert into locations values(?,?,?)'
```

```python
#데이터 적재 SQL 수행 
# 데이터의 개수만큼 수행하기 위해 executemany() 함수를 사용
# 첫 번째 인자 : SQL 구문(insert into)
# 두 번째 인자 : 삽입하고자 하는 데이터
conn.executemany(삽입, data)
```

```python
# commit() 수행 
conn.commit()
```

```python
#Select SQL 실행1
# pd.read_sql() 사용
# 첫 번째 인자 : [select] 구문
# 두 번째 인자 : [conn] 커넥션 인자
pd.read_sql('select * from locations', conn)
```

```python
#Select SQL 실행2
pd.read_sql('select a, sum(c) from locations group by a', conn)
```

```python
# 크기가 큰 테이블의 데이터 읽어오기
# 데이터베이스의 테이블 크기가 큰 경우에는 cursor 사용하기
# Cursor를 활용하여 데이터 끊어서 읽어오기
# cursor : execute() 함수에 select 구문을 인자로 전달하면, select 구문을 실행할 준비가 된 Cursor 객체를 반환함
cursor = conn.execute('select * from locations')
cursor
```

```python
# cursor.fetchmany(cnt) : cnt 개수만큼 데이터를 읽어옴
# cnt 개수만큼 데이터 끊어서 읽어 옴
cursor.fetchmany(2)
```

```python
# cursor나 데이터베이스의 conn(커넥션) 같은 경우 close() 를 해 주는 게 기본적인 자원 관리를 위해 필요함
cursor.close()
conn.close()
```
