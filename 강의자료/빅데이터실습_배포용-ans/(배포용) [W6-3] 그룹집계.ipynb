```python
import pandas as pd
from pandas import Series, DataFrame
```

```python
# 데이터 분석하기
# 그룹집계(Group Aggregate)
# 1) 그룹집계의 개념
# 2) groupby() 함수를 활용한 그룹집계
# 3) pivot_table() 함수를 활용한 그룹집계
```

```python
# 그룹집계
# 데이터를 특정 속성의 값으로 그룹핑하여 집계를 수행함
# 특정 컬럼의 값을 기준으로 데이터를 그룹핑하여 그룹별로 집계하는 방식은 많이 활용된다.(e.g. sum, max, min, mean, std 등)
# 전체 판매량을 구하려면?
# data.sum()
# 제품별 판매량을 구하려면?
# data[data.제품명 == 'A'].sum()
# data[data.제품명 == 'B'].sum()
# data[data.제품명 == 'C'].sum()
```

```python
# SQL 구문의 GROUP BY 연산과 유사
# 그룹 분석을 하는 방법
# 1) group by
# groupby 컬럼의 값으로 데이터들을 그룹핑함
# 집계 함수 수행(e.g. sum)
# 집계 함수 적용(sum, mean, max, min, std, var 등 연산 가능)
# 복수 개의 컬럼 지정도 가능, 리스트로 나열
# data.groupby('제품명')['판매량'].sum()
```

```python
# 실습 데이터 적재
# 도시의 자치구별, 연도별, 남자인구와 여자인구 그리고 총인구
data = pd.read_excel('data/인구수예제.xlsx')
data.head()
```

```python
# 실습 1. 연도별 총인구수 구하기
# 연도별로 그룹핑
# groupby 인자에 그룹핑할 기준 컬럼인 '연도' 컬럼을 지정
# 총인구 컬럼의 합(sum) 수행
data.groupby('연도')['총인구'].sum()
```

```python
# 실습 2. 연도별 전체의 남자인구, 여자인구, 총인구 수 구하기
# 연도별로 그룹핑 후 남자인구, 여자인구, 총인구를 리스트로 전달
data.groupby('연도')[['남자인구', '여자인구', '총인구']].sum()
```

```python
# 실습 3. 자치구별로 평균 총인구수 구하기
# 자치구별로 그룹핑 후, 총인구수의 평균 구하기
# 같은 자치구의 이름이 다른 도시에 있을 수 있기 때문에 도시와 자치구별로 구해야 함
# groupby 인자에 도시, 자치구를 리스트로 담아서 전달
data.groupby(['도시', '자치구'])['총인구'].mean()
```

```python
# 실습 4.도시/자치구별 평균 총인구수 구하기
data.groupby(['도시', '자치구'])['총인구'].mean()
```

```python
# 실습 5. 도시/자치구별 평균 남자인구와 여자인구수 구하기
# 남자인구, 여자인구를 리스트로 담아 전달
# groupby의 columns과 집계하고자 하는 columns을 리스트로 담아 여러 개를 전달할 수 있음
# 집계 함수를 동시에 여러 개 사용할 수는 없음
data.groupby(['도시', '자치구'])[['남자인구', '여자인구']].mean()
```

```python
# 실습 6. 도시별, 연도별로 총인구수 출력
# groupby 함수에 도시, 연도를 인자로 담고, 총인구의 합을 구함
# columns index 계층 색인으로 변경
# unstack에 레벨 0또는 '도시'명명
data.groupby(['도시', '연도'])['총인구'].sum().unstack('도시')
```

```python
# groupby : 집계 결과를 보기 좋게 하기 위해 stack이나 unstack 등의 구문을 추가로 작성해야 되는 경우가 있음
# pivot_table : 매우 직관적임
```

```python
# 실습 2. 연도별 전체의 남자인구, 여자인구, 총인구 수와 평균 구하기
# pivot_table() : groupby 컬럼이 1개(연도)이므로, index나 columns 인자 중 하나만 사용
# 집계 함수를 2개 이상 사용 가능(리스트 형태)
data.pivot_table(index = '연도', aggfunc = ['sum', 'mean'], values = ['남자인구', '여자인구', '총인구'])
```

```python
# 깔끔하게 도시별, 연도별 총인구의 sum을 구할 수 있음
# 실습 6. 도시별, 연도별로 총인구수 출력
data.pivot_table(index = '연도', columns = '도시', aggfunc = 'sum', values = '총인구')
```

```python
# 실습 7. 자치구 별로 남자인구의 평균을 구한 후, 남자 인구가 가장 많은 도시 및 자치구 Top 5 찾기
# 평균 구하기
# index : ['도시', '자치구']
# columns : X
# 남자인구에 평균(mean) 적용
# Top 5 찾기(정렬)
# sort_values : 내림차순으로 남자인구 정렬하고 head(5)
# 역슬래스(\)는 하나의 구문을 줄바꿈할 때 사용하며, 구문이 콤마로 구분되는 경우에는 역슬래시(\) 없이 줄바꿈 가능
data.pivot_table(index = ['도시', '자치구'], aggfunc = 'mean', values = '남자인구')\
                .sort_values(by = '남자인구', ascending = False).head()
```

```python
# 실습 8. 여자가 남자보다 많은 도시/자치구 상위 3개 찾기 
# 여자가 남자보다 많은 도시 찾기(5년 평균)
# 여자와 남자 인구 차이를 계산하여 새로운 columns(남녀차이) 추가
# + : 여자 인구가 많은 곳, - : 여자 인구가 적은 곳
data['남녀차이'] = data.여자인구 - data.남자인구
data
```

```python
# index : 도시, 자치구
# 남녀차이 columns의 평균 구하기
# - : 남자 인구가 많은 곳, + : 여자 인구가 많은 곳
# 상위 3개 찾기(정렬)
# sort_values를 이용하여 남녀차이로 내림차순하여 head(3)
data.pivot_table(index = ['도시', '자치구'], aggfunc = 'mean', values = '남녀차이')\
.sort_values(by = '남녀차이', ascending = False).head(3)
```
